```python
import pandas as pd
from IPython.display import display
import seaborn as sns
import matplotlib.pyplot as plt
import datetime


df = pd.read_csv("C:/Users/saumy/Downloads/indusbank/INDUSINDBK.csv")
display(df)
```

**Condition for extraction o stock price of financial year 2010**

```python
#Condition for extraction o stock price of financial year 2010
df['Date'] = pd.to_datetime(df['Date'])

start_date = '2010-04-01'
end_date = '2011-03-31'

stockprice_one_year = (df['Date'] > start_date) & (df['Date'] <= end_date)
  
stockprice_one_year = df.loc[stockprice_one_year]
display(stockprice_one_year)
```

**Function for Converting date yyyy-mm-dd to dd-mm-yyyy**

```python
def conv_dates_series(df, col, odf, ndf):
    df[col] = pd.to_datetime(df[col], format=odf).dt.strftime(ndf)
    return df

odf='%Y-%m-%d'
ndf='%d/%m/%Y'

# display(conv_dates_series(df, "Date", odf, ndf))
```

**Compare stockprice_one_year['feature_name'].plot() vs df['feature_name'].plot()**

```python
plt.subplot(2, 1, 1)
plt.plot(stockprice_one_year['Date'],stockprice_one_year['Open'])


plt.subplot(2, 1, 2)
plt.plot(df['Date'],df['Open'])
plt.suptitle("Opening Value Comparison")

plt.show()
```

**Check whether matplotlib's plotting is better in looks. Once the plotting is done, check what is the impact in Plots, if you execute the line of code: plt.style.use('fivethirtyeight')**

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8,6))
    
# Plotting each occupation category
ax1 = sns.lineplot(x=df['Date'], y=df['Open'].tolist(), color='#0F95D7', lw=2.5)
ax2 = sns.lineplot(x=df['Date'], y=df['High'].tolist(), color='#FF2700', lw=2.5)
ax3 = sns.lineplot(x=df['Date'], y=df['Low'].tolist(), color='#810F7C', lw=2.5)
```

**Find the correlation between the features. Make a Correlation Heatmap. Write your observations about the same.**

```python
print("                          Correlation Of Dataframe")
print("-"*77)
display(df.corr())
print("_"*77)
corrdf = df.corr()
print("                         Heatmap of The Corellation")
print("-"*77)
hm = sns.heatmap(corrdf)
```

**How can you plot multiple features in a single plot? For instance, use a red line for "LOW", green line for "HIGH", and black line for "CLOSE".**

```python
fig, ax = plt.subplots(figsize=(8,6))

ax1 = sns.lineplot(x=df['Date'], y=df['Low'].tolist(), color='#FF0000', lw=2.5)
ax2 = sns.lineplot(x=df['Date'], y=df['High'].tolist(), color='#00FF00', lw=2.5)
ax3 = sns.lineplot(x=df['Date'], y=df['Close'].tolist(), color='#000000', lw=2.5)
```

**What does plt.fill_between() do? can you shade the region between LOW and HIGH?**

```python
x = df['Low']
y = df['High']


plt.plot(x, y,'-o', color ='gray')
  
# Filling between line a3 
# and line a4

plt.fill_between(x,y)
plt.show()
```
